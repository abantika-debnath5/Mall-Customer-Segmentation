```python
import mysql.connector
conn = mysql.connector.connect(
    host = "127.0.0.1",
    user = "root",
    password = "2005",
    database = "mall_project2")
cursor = conn.cursor()
print(cursor)
```

```python
query = """
SELECT
    CASE
        WHEN annual_income < 40 THEN 'Low Income'
        WHEN annual_income BETWEEN 40 AND 70 THEN 'Mid Income'
        ELSE 'High Income'
    END AS income_group,
    CASE
        WHEN spending_score < 40 THEN 'Low Spending'
        WHEN spending_score BETWEEN 40 AND 70 THEN 'Medium Spending'
        ELSE 'High Spending'
    END AS spending_group,
    COUNT(*) AS customers
FROM mall_customers
GROUP BY income_group, spending_group;
"""
cursor.execute(query)
rows = cursor.fetchall()
print(rows)
```

```python
segment_counts = {}

for income, spending, count in rows:
    key = f"{income} - {spending}"
    segment_counts[key] = count
print(segment_counts)
```

```python
total_customers = sum(segment_counts.values())

segment_percentages = {}

for segment, count in segment_counts.items():
    segment_percentages[segment] = round((count / total_customers) * 100, 2)
print(segment_percentages)
```

```python
sorted_segments = sorted(
    segment_percentages.items(),
    key=lambda x: x[1],
    reverse=True
)

for segment, pct in sorted_segments:
    print(segment, ":", pct, "%")
print(sorted_segments)
```
